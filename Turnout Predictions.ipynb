```python
def get_total_num_registered_by_precinct(precinct, registration_data, registration_year):
    registration_by_precinct = registration_data.groupby('Precinct')
    precinct_reg = registration_by_precinct.get_group(precinct)
    total_registered_in_precinct = sum(precinct_reg['Total Registered: '+str(registration_year)])
    return [registration_by_precinct, total_registered_in_precinct]
```

```python
def get_num_and_percent_increase_registration(total_2020, total_2022):
    num_registration_increase = total_2022 - total_2020
    perc_registration_increase = num_registration_increase / total_2020 * 100
    return [num_registration_increase, perc_registration_increase]
```

```python
# accounting for edge cases of no data of that party

def get_num_percent_each_party_registered_by_precinct(precinct, registration_data, registration_year):
    registration_by_precinct = get_total_num_registered_by_precinct(precinct, registration_data, registration_year)[0]
    total_registered_in_precinct = get_total_num_registered_by_precinct(precinct, registration_data, registration_year)[1]
    precinct_reg = registration_by_precinct.get_group(precinct)
    precinct_parties = precinct_reg.groupby('Party')
    
    
    if 'Democrat' in precinct_reg.Party.values:
        democrat_data = precinct_parties.get_group('Democrat')
        total_num_democrat = sum(democrat_data['Total Registered: '+str(registration_year)])
        percent_democrat = int(total_num_democrat / total_registered_in_precinct * 100)
        
    else:
        total_num_democrat = 0
        percent_democrat = 0
    
    if 'Republican' in precinct_reg.Party.values:
        republican_data = precinct_parties.get_group('Republican')
        total_num_republican = sum(republican_data['Total Registered: '+str(registration_year)])
        percent_republican = int(total_num_republican / total_registered_in_precinct * 100)
    
    else:
        total_num_republican = 0
        percent_republican = 0
        
    if 'Independent / Undecided / Swing' in precinct_reg.Party.values:
        indep_swing_undecided_data = precinct_parties.get_group('Independent / Undecided / Swing')
        total_num_indep_swing_undecided = sum(indep_swing_undecided_data['Total Registered: '+str(registration_year)])
        percent_indep_swing_undecided = int(total_num_indep_swing_undecided / total_registered_in_precinct * 100)
    
    else:
        total_num_indep_swing_undecided = 0
        percent_indep_swing_undecided = 0
    
    if 'No Data / Unknown' in precinct_reg.Party.values:
        no_data_unknown_data = precinct_parties.get_group('No Data / Unknown')
        total_num_no_data_unknown = sum(no_data_unknown_data['Total Registered: '+str(registration_year)])
        percent_no_data_unknown = int(total_num_no_data_unknown / total_registered_in_precinct * 100)
    
    else:
        total_num_no_data_unknown = 0
        percent_no_data_unknown = 0
    
    num_percent_each_party = {'Party': ['Democrat', 'Republican', 'Independent / Undecided / Swing', 'No Data / Unknown'], 
    'Total Registered in Party: '+str(registration_year): [total_num_democrat, total_num_republican, total_num_indep_swing_undecided, total_num_no_data_unknown],
    '% of Total Registration - '+str(registration_year): [percent_democrat, percent_republican, percent_indep_swing_undecided, percent_no_data_unknown]}
    
    num_percent_each_party = pd.DataFrame(num_percent_each_party)
    
    return num_percent_each_party
```

```python
# accounting for edge cases

def get_num_percent_each_race_registered_by_precinct(precinct, registration_data, registration_year):
    registration_by_precinct = get_total_num_registered_by_precinct(precinct, registration_data, registration_year)[0]
    total_registered_in_precinct = get_total_num_registered_by_precinct(precinct, registration_data, registration_year)[1]
    precinct_reg = registration_by_precinct.get_group(precinct)
    precinct_races = precinct_reg.groupby('Race')
    
    if 'White' in precinct_reg.Race.values:
        white_data = precinct_races.get_group('White')
        total_num_white = sum(white_data['Total Registered: '+str(registration_year)])
        percent_white = total_num_white / total_registered_in_precinct * 100
    
    else:
        total_num_white = 0
        percent_white = 0
    
    if 'Asian' in precinct_reg.Race.values:
        asian_data = precinct_races.get_group('Asian')
        total_num_asian = sum(asian_data['Total Registered: '+str(registration_year)])
        percent_asian = total_num_asian / total_registered_in_precinct * 100
    
    else:
        total_num_asian = 0
        percent_asian = 0
        
    
    if 'Hispanic' in precinct_reg.Race.values:
        hispanic_data = precinct_races.get_group('Hispanic')
        total_num_hispanic = sum(hispanic_data['Total Registered: '+str(registration_year)])
        percent_hispanic = total_num_hispanic / total_registered_in_precinct * 100
    
    else:
        total_num_hispanic = 0
        percent_hispanic = 0
        
    
    if 'Black' in precinct_reg.Race.values:
        black_data = precinct_races.get_group('Black')
        total_num_black = sum(black_data['Total Registered: '+str(registration_year)])
        percent_black = total_num_black / total_registered_in_precinct * 100
    
    else:
        total_num_black = 0
        percent_black = 0
        
    
    if 'Unknown' in precinct_reg.Race.values:
        unknown_data = precinct_races.get_group('Unknown')
        total_num_unknown = sum(unknown_data['Total Registered: '+str(registration_year)])
        percent_unknown = total_num_unknown / total_registered_in_precinct * 100
    
    else:
        total_num_unknown = 0
        percent_unknown = 0
    
    num_percent_each_race = {'Race': ['White', 'Asian', 'Hispanic', 'Black', 'Unknown'], 
    'Total Registered: '+str(registration_year): [total_num_white, total_num_asian, total_num_hispanic, total_num_black, total_num_unknown],
    '% of Total Registration - '+str(registration_year): [percent_white, percent_asian, percent_hispanic, percent_black, percent_unknown]}
    
    num_percent_each_race = pd.DataFrame(num_percent_each_race)
    
    return num_percent_each_race
```

```python
def get_turnout_and_prediction_by_race_in_precinct(precinct, registration_2020_with_race, registration_2022_with_race, voter_2020_data_with_race):
    registration_2020_by_race = get_num_percent_each_race_registered_by_precinct(precinct, registration_2020_with_race, 2020)
    registration_2022_by_race = get_num_percent_each_race_registered_by_precinct(precinct, registration_2022_with_race, 2022)
    voters_2020_by_race = get_num_percent_each_race_voters_by_precinct(precinct, voters_2020_with_race)
    registration_voter_data_2020_by_race = pd.merge(registration_2020_by_race, voters_2020_by_race)
    registration_voter_data_2020_by_race['% Turnout - 2020'] = registration_voter_data_2020_by_race['Total Voters: 2020'] / registration_voter_data_2020_by_race['Total Registered: 2020'] * 100
    turnout_and_prediction_by_race = pd.merge(registration_voter_data_2020_by_race, registration_2022_by_race)
    turnout_and_prediction_by_race['% Registration Increase 2020 - 2022'] = turnout_and_prediction_by_race['% of Total Registration - 2022'] - turnout_and_prediction_by_race['% of Total Registration - 2020']
    turnout_and_prediction_by_race['Projected Number of Voters: 2022'] = turnout_and_prediction_by_race['% Turnout - 2020'] / 100 * turnout_and_prediction_by_race['Total Registered: 2022']
    turnout_and_prediction_by_race.dropna(inplace=True)
    turnout_and_prediction_by_race['Projected Number of Voters: 2022'] = turnout_and_prediction_by_race['Projected Number of Voters: 2022'].astype('int')
    return turnout_and_prediction_by_race
```

```python
# merges registration for 2020 & 2022, voter data for 2020 to get turnout + prediction by age / gender

def get_turnout_and_prediction_by_age_and_gender_in_precinct(precinct, registration_2020_with_race, registration_2022_with_race, voter_2020_data_with_race):
    registration_2020_by_age_gender = get_num_percent_age_gender_registered_by_precinct(precinct, registration_2020_with_race, 2020)
    registration_2022_by_age_gender = get_num_percent_age_gender_registered_by_precinct(precinct, registration_2022_with_race, 2022)
    voters_2020_by_age_gender = get_num_percent_age_gender_voters_by_precinct(precinct, voters_2020_with_race)
    registration_voter_data_2020_by_age_gender = pd.merge(registration_2020_by_age_gender, voters_2020_by_age_gender)
    registration_voter_data_2020_by_age_gender['% Turnout - 2020'] = registration_voter_data_2020_by_age_gender['Total Voters: 2020'] / registration_voter_data_2020_by_age_gender['Total Registered: 2020'] * 100
    turnout_and_prediction_by_age_gender = pd.merge(registration_voter_data_2020_by_age_gender, registration_2022_by_age_gender)
    turnout_and_prediction_by_age_gender['% Registration Increase 2020 - 2022'] = turnout_and_prediction_by_age_gender['% of Total Registration - 2022'] - turnout_and_prediction_by_age_gender['% of Total Registration - 2020']
    turnout_and_prediction_by_age_gender['Projected Number of Voters: 2022'] = turnout_and_prediction_by_age_gender['% Turnout - 2020'] / 100 * turnout_and_prediction_by_age_gender['Total Registered: 2022']
    turnout_and_prediction_by_age_gender.dropna(inplace=True)
    turnout_and_prediction_by_age_gender['Projected Number of Voters: 2022'] = turnout_and_prediction_by_age_gender['Projected Number of Voters: 2022'].astype('int')
    return turnout_and_prediction_by_age_gender
```

```python
# district breakdown by party: merged, turnout + prediction
def get_turnout_and_prediction_by_party_in_district(registration_2020_with_party, registration_2022_with_party, voter_2020_data_with_party):
    registration_2020_by_party = get_num_percent_each_party_registered_in_district(registration_2020_with_party, 2020)
    registration_2022_by_party = get_num_percent_each_party_registered_in_district(registration_2022_with_party, 2022)
    voters_2020_by_party = get_num_percent_each_party_voters_in_district(voter_2020_data_with_party)
    registration_voter_data_2020_by_party = pd.merge(registration_2020_by_party, voters_2020_by_party)
    registration_voter_data_2020_by_party['% Turnout - 2020'] = registration_voter_data_2020_by_party['Total Voters in Party: 2020'] / registration_voter_data_2020_by_party['Total Registered in Party: 2020'] * 100
    turnout_and_prediction_by_party = pd.merge(registration_voter_data_2020_by_party, registration_2022_by_party)
    turnout_and_prediction_by_party['% Registration Increase 2020 - 2022'] = turnout_and_prediction_by_party['% of Total Registration - 2022'] - turnout_and_prediction_by_party['% of Total Registration - 2020']
    turnout_and_prediction_by_party['Projected Number of Voters: 2022'] = turnout_and_prediction_by_party['% Turnout - 2020'] / 100 * turnout_and_prediction_by_party['Total Registered in Party: 2022']
    turnout_and_prediction_by_party.dropna(inplace=True)
    turnout_and_prediction_by_party['Projected Number of Voters: 2022'] = turnout_and_prediction_by_party['Projected Number of Voters: 2022'].astype('int')
    return turnout_and_prediction_by_party
```
